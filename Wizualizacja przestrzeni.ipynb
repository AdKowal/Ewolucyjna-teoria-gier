```python
import matplotlib.pyplot as plt

# Utwórz macierz 9x9 z zerami
grid = [[0 for x in range(9)] for y in range(9)]

# Wypełnij macierz wartościami od 1 do 9
for i in range(9):
    for j in range(9):
        grid[i][j] = (i * 3 + i // 3 + j) % 9 + 1

# Utwórz wykres
fig, ax = plt.subplots()

# Ustaw zakres osi x i y na [0, 3] i wyłącz osie
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.axis('off')

# Podziel kwadrat na 9 sekcji
for i in range(3):
    for j in range(3):
        # Oblicz współrzędne środka sekcji
        x = i + 0.5
        y = j + 0.5
        
        # Umieść liczbę w środku sekcji
        ax.text(x, y, str(grid[i*3][j*3]), ha='center', va='center', fontsize=20)
        ax.text(x+1, y, str(grid[i*3][j*3+1]), ha='center', va='center', fontsize=20)
        ax.text(x+2, y, str(grid[i*3][j*3+2]), ha='center', va='center', fontsize=20)
        ax.text(x, y+1, str(grid[i*3+1][j*3]), ha='center', va='center', fontsize=20)
        ax.text(x+1, y+1, str(grid[i*3+1][j*3+1]), ha='center', va='center', fontsize=20)
        ax.text(x+2, y+1, str(grid[i*3+1][j*3+2]), ha='center', va='center', fontsize=20)
        ax.text(x, y+2, str(grid[i*3+2][j*3]), ha='center', va='center', fontsize=20)
        ax.text(x+1, y+2, str(grid[i*3+2][j*3+1]), ha='center', va='center', fontsize=20)
        ax.text(x+2, y+2, str(grid[i*3+2][j*3+2]), ha='center', va='center', fontsize=20)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Tworzenie figury
fig, ax = plt.subplots()

# Rysowanie linii pionowych
ax.axvline(x=1/3, color='black')
ax.axvline(x=2/3, color='black')

# Rysowanie linii poziomych
ax.axhline(y=1/3, color='black')
ax.axhline(y=2/3, color='black')

# Ustawienie limitów osi
ax.set_xlim([0,1])
ax.set_ylim([0,1])

# Ukrycie osi
ax.axis('off')

# Wyświetlenie rysunku
plt.show()
```
