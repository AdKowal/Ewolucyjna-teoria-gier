```python
import pandas as pd
import numpy as np
import random
import time
```

```python
n = 100
t = 20
```

```python
s = []
for i in range(n):
    s.append(random.choice(["K", "P", "E", "R"]))
p = []
for i in range(n):
    p.append(random.choice(["B", "D"]))
ż = []
for i in range(n):
    ż.append(random.choice(["Z", "K"]))
f = []
for i in range(n):
    f.append(random.choice(["A","B"]))
Strategia = pd.DataFrame(s, columns=["Strategia"])
Pamięć = pd.DataFrame(p, columns = ["Pamięć"])
Fenotyp = pd.DataFrame(f, columns=["Fenotyp"])
Tryb = pd.DataFrame(ż, columns=["Tryb życia"])
SiF = pd.merge(Strategia, Pamięć, right_index=True, left_index=True, how="inner")
SiF = pd.merge(SiF, Fenotyp, right_index=True, left_index=True, how="inner")
SiF = pd.merge(SiF, Tryb, right_index=True, left_index=True, how="inner")
```

```python
#table_1 pandas
SiF1=SiF
SiF1["Punkty"] = 100
SiF1["Wiek"]=0
SiF1["Status"] = "Ż"
SiF1.loc[SiF1["Fenotyp"] == "A", "Fenotyp A"] = 100
SiF1.loc[SiF1["Fenotyp"] == "B", "Fenotyp B"] = 100
SiF1.fillna(value=0, inplace=True)
l = []
for i in range(n):
    l.append(random.randrange(1,5))
Lokalizacja = pd.DataFrame(l, columns=["Lokalizacja"])
SiF1 = pd.merge(SiF1, Lokalizacja, right_index=True, left_index=True, how="inner")
Gra=SiF1.drop(["Fenotyp"], axis = 1)
Gra
```

```python
#table_2 numpy
Gra2 = np.zeros((n, 1))

index = np.arange(100)
Gra2[:,0] = index

strategia = np.random.randint(4, size=(n, 1))
Gra2 = np.hstack((Gra2, strategia))

pamięć = np.random.randint(2, size=(n, 1))
Gra2 = np.hstack((Gra2, pamięć))

tryb_życia= np.random.randint(2, size=(n, 1))
Gra2 = np.hstack((Gra2, tryb_życia))

pkt = np.full_like(Gra2[:, :1], 100)
Gra2 = np.hstack((Gra2, pkt))

wiek = np.full_like(Gra2[:, :1], 0)
Gra2 = np.hstack((Gra2, wiek))

status = np.full_like(Gra2[:, :1], 0)
Gra2 = np.hstack((Gra2, status))

Fenotyp = np.random.randint(2, size=(n, 1))
Gra2 = np.hstack((Gra2, Fenotyp))

FenotypA = np.full_like(Gra2[:, :1], 0)
Gra2 = np.hstack((Gra2, FenotypA))
FenotypA = np.where(Gra2[:,7] == 0, 100, Gra2[:, 8])
Gra2[:,8] = FenotypA

FenotypB = np.full_like(Gra2[:, :1], 0)
Gra2 = np.hstack((Gra2, FenotypB))
FenotypB = np.where(Gra2[:,7] == 1, 100, Gra2[:, 9])
Gra2[:,9] = FenotypB

Lokalizacja = np.random.randint(4, size=(n, 1))
Gra2 = np.hstack((Gra2, Lokalizacja))
Gra2 = np.delete(Gra2, 7, axis=1)
Gra2
```

```python
def stg():
    global g2
    if Gra.loc[j, "Strategia"] == "K":
        g2 = 1
    elif Gra.loc[j, "Strategia"] == "E":
        g2 = 0
    elif Gra.loc[j, "Strategia"] == "P":
        if Gra.loc[[j], ["Fenotyp A"]].equals(Gra.loc[[j], ["Fenotyp B"]]):
            g2 = 1
        else:
            N_j2 = Gra.loc[[j], ["Fenotyp A", "Fenotyp B"]].idxmax(axis = 1).item()
            N_i2 = Gra.loc[[i], ["Fenotyp A", "Fenotyp B"]].idxmax(axis = 1).item()
            if N_j2 == N_i2:
                g2 = 1
            else:
                g2 = 0
    elif Gra.loc[j, "Strategia"] == "R":
        Z_j2 = Gra.loc[[j], ["Fenotyp A", "Fenotyp B"]].idxmax(axis = 1).item()
        Z_i2 = Gra.loc[[i], ["Fenotyp A", "Fenotyp B"]].idxmax(axis = 1).item()
        if Z_j2 != Z_i2:
            g2 = 1
        else:
            g2 = 0
```

```python
start_time = time.time()
for _ in range(t):
    for i in range(0, len(Gra)):
        if Gra.loc[i, "Status"] == "Ś":
            continue
        for j in range(0, i):
            if Gra.loc[j, "Status"] == "Ś":
                continue
            if Gra.loc[i, "Status"] == "Ś":
                continue
            if Gra.loc[i, "Lokalizacja"] != Gra.loc[j, "Lokalizacja"]:
                continue
            if Gra.loc[i, "Strategia"] == "K":
                g1 = 1
                stg()
            elif Gra.loc[i, "Strategia"] == "E":
                g1 = 0
                stg()
            elif Gra.loc[i, "Strategia"] == "P":
                if Gra.loc[[i], ["Fenotyp A"]].equals(Gra.loc[[i], ["Fenotyp B"]]):
                    g1 = 1
                else:    
                    N_i1 = Gra.loc[[i], ["Fenotyp A", "Fenotyp B"]].idxmax(axis = 1).item()
                    N_j1 = Gra.loc[[j], ["Fenotyp A", "Fenotyp B"]].idxmax(axis = 1).item()
                    if N_j1 == N_i1:
                        g1 = 1
                    else:
                        g1 = 0
                stg()
            elif Gra.loc[i, "Strategia"] == "R":
                Z_j1 = Gra.loc[[j], ["Fenotyp A", "Fenotyp B"]].idxmax(axis = 1).item()
                Z_i1 = Gra.loc[[i], ["Fenotyp A", "Fenotyp B"]].idxmax(axis = 1).item()
                if Z_j1 != Z_i1:
                    g1 = 1
                else:
                    g1 = 0
                stg()
            if g1 == 1:
                if g2 == 1:
                    Gra.loc[i, "Punkty"] += 3
                    Gra.loc[j, "Punkty"] += 3
                else:
                    Gra.loc[i, "Punkty"] += 0
                    Gra.loc[j, "Punkty"] += 5
            else:
                if g2 == 1:
                    Gra.loc[i, "Punkty"] += 5
                    Gra.loc[j, "Punkty"] += 0
                else:
                    Gra.loc[i, "Punkty"] += 1
                    Gra.loc[j, "Punkty"] += 1
    print(f"Runda {_}")
    Gra["Wiek"] += 1
end_time = time.time()
execution_time_1 = end_time - start_time
print("Czas wykonania pętli: ", execution_time_1, "sekundy")
```
